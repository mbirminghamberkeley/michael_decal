```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns 

def make_sin_wave(x, A, w):
    return A*np.sin(w*x)

x = np.arange(0, 2*np.pi, np.pi/500)

amp = np.array([0.5,1,1.5,2,2.5], dtype=float)
freq = np.array([1,2,3,4,5], dtype=int)

plt.figure(figsize=(8,6))

both = zip(amp, freq)
for a, w in both:
    y_values = make_sin_wave(x, a, w)
    plt.plot(x, y_values, label = f"amplitude: {str(a)}, frequency: {str(w)}")

plt.xlabel("x")
plt.ylabel("sin x")
plt.grid()
plt.legend()
plt.title("Sin x vs x for various frequencies and amplitudes")
plt.show()

```
